# Funcional

```python
# s0
import csv
import pandas as pd
```

```python
# s1
def load_csv(filepath: str, delimiter: str, header='infer', encoding='utf-8'):
    return pd.read_csv(filepath, delimiter=delimiter, header=header, encoding=encoding)   
```

```python
df = load_csv(filepath=file, delimiter=';') # cqd
df.head()
```

```python
# s2
def find_ids(df, id_column: str):
    return list(set(df[id_column]))
```

```python
idArray = find_ids(df=df, id_column='MATRICULA_FUNCIONARIO') # cqd
# idArray
```

```python
# s3
def single_id_df (df, id_column: str, id_value: any):
    return df[df[id_column] == id_value]
```

```python
df_id0 = single_id_df(df, id_column='MATRICULA_FUNCIONARIO', id_value=idArray[0]) # cqd
df_id0
```

```python
# s4
def extract_features_by_category(single_id_df, category: str, related_features: list):
    lc = [ [str(reg[i]).rstrip().lower() for i in range(len(reg)) ] 
          for reg in single_id_df[single_id_df['TIPO'] == category][related_features].values ]
    related_features = [feat.lower() for feat in related_features]
    out = []
    for reg in lc:
        out.append(dict(zip(related_features, reg)))
    category_key = category.lower()
    category_dict = {category_key: out}
    return category_dict
```

```python
extract_features_by_category(single_id_df=df_id0,
                         category='REALIZACAO', related_features=['CODIGO', 'NOME']) # cqd
```

```python
colsFormacao = ['CODIGO', 'NOME', 'NIVEL', 'DATA_FIM',
             'MODALIDADE', 'NATUREZA', 'URL', 'ESTADO',
             'NOME_INSTITUICAO_ENSINO']
extract_features_by_category(single_id_df=df_id0,
                            category='FORMACAO_SUPERIOR',
                            related_features=colsFormacao) # cqd
```

```python
mycsv = '/home/f4119597/Downloads/peopleAnalytics/ListaGeralDitec-VRS20191001/QueryDB2TAO-GERAL-DITEC-VRS20191001.csv'
```

```python
df = load_csv(filepath=mycsv, delimiter=';', header='infer', encoding='cp1252')
```

```python
df.columns
```

```python
idList = find_ids(df=df, id_column='MATRICULA_FUNCIONARIO')
```

```python
idList[:5]
```

```python
campos = ['CAPACIDADE', 'CONHECIMENTO', 'FERRAMENTA', 'INTERESSE',
          'MOTIVACAO', 'REALIZACAO', 'FORMACAO_SUPERIOR']

rltd_capacidade = ['NOME', 'ATIVO', 'CODIGO']
rltd_conhecimento = ['NOME', 'CODIGO', 'NIVEL', 'CODIGO_AREA', 'CODIGO_SUB_AREA']
rltd_ferramenta = ['NOME', 'CODIGO', 'ATIVO']
rltd_interesse = ['NOME', 'CODIGO']
rltd_motivacao = ['NOME', 'CODIGO']
rltd_realizacao = ['NOME', 'CODIGO']
rltd_form_superior = ['NOME', 'CODIGO', 'NIVEL', 'DATA_FIM',
                        'MODALIDADE', 'NATUREZA', 'URL', 'ESTADO',
                        'NOME_INSTITUICAO_ENSINO']

rltds = [rltd_capacidade, rltd_conhecimento, rltd_ferramenta, rltd_interesse,
        rltd_motivacao, rltd_realizacao, rltd_form_superior]

targets = dict(zip(campos, rltds))
```

```python
targets
```

```python
out = []

for f_id in idList[:5]:
    
    f_info = {'id': str(f_id), 'curriculo': {} }
    
    f_df = single_id_df(df=df, id_column='MATRICULA_FUNCIONARIO', id_value=f_id)
    
    #
    capacidade = extract_features_by_category(single_id_df=f_df,
                                              category='CAPACIDADE',
                                              related_features=rltd_capacidade)
    f_info['curriculo']['capacidade'] = capacidade
    
    #
    conhecimento = extract_features_by_category(single_id_df=f_df,
                                              category='CONHECIMENTO',
                                              related_features=rltd_conhecimento)
    f_info['curriculo']['conhecimento'] = conhecimento
    
    #
    ferramenta = extract_features_by_category(single_id_df=f_df,
                                              category='FERRAMENTA',
                                              related_features=rltd_ferramenta)
    f_info['curriculo']['ferramenta'] = ferramenta
    
    #
    interesse = extract_features_by_category(single_id_df=f_df,
                                              category='INTERESSE',
                                              related_features=rltd_interesse)
    f_info['curriculo']['interesse'] = interesse
    
    #
    motivacao = extract_features_by_category(single_id_df=f_df,
                                              category='MOTIVACAO',
                                              related_features=rltd_motivacao)
    f_info['curriculo']['motivacao'] = motivacao
    
    #
    realizacao = extract_features_by_category(single_id_df=f_df,
                                              category='REALIZACAO',
                                              related_features=rltd_realizacao)
    f_info['curriculo']['realizacao'] = realizacao
    
    #
    f_superior = extract_features_by_category(single_id_df=f_df,
                                              category='FORMACAO_SUPERIOR',
                                              related_features=rltd_form_superior)
    f_info['curriculo']['f_superior'] = f_superior
    
    out.append(f_info)
    
```

```python
out = []
for f_id in idList:
    f_info = {'id': str(f_id), 'curriculo': {} }
    f_df = single_id_df(df=df, id_column='MATRICULA_FUNCIONARIO', id_value=f_id)
    for key in targets.keys():
        try:
            f_info['curriculo'][key.lower()] = extract_features_by_category(single_id_df=f_df,
                                                              category=key,
                                                              related_features=targets[key])
        except:
            print('ERRO! id \'{}\' e campo \'{}\''.format(f_id, key))
        out.append(f_info)
```
